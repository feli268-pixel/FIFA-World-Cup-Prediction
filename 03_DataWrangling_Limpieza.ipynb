```python
import pandas as pd

partidos=pd.read_csv("todos_los_mundiales.csv")
fixtures=pd.read_csv("fixtures.csv")

fixtures["Local"]=fixtures["Local"].str.strip() #con .strip() Limpiamos espacios en blanco,saltos de linea etc
fixtures["Visitante"]=fixtures["Visitante"].str.strip()
```

```python
partidos["Local"]=partidos["Local"].str.strip()
partidos["Visitante"]=partidos["Visitante"].str.strip()
```

```python
partidos.dropna(inplace=True)#Elimina datos faltantes (NaN)
partidos[partidos["Marcador"].isnull()]
partidos.drop_duplicates(inplace=True)#Eliminamos duplicados
index_eliminar=partidos[partidos["Local"].str.contains("Sweden") & partidos["Visitante"].str.contains("Austria")].index
partidos.drop(index=index_eliminar,inplace=True)
```

```python
partidos[partidos["Marcador"].str.contains("[^\d-]")]
partidos["Marcador"]=partidos["Marcador"].str.replace("[^\d–]","",regex=True)
partidos[partidos["Marcador"].str.contains("[^\d-]")]
partidos
```

```python
partidos[["Goles Local","Goles Visitante"]]=partidos["Marcador"].str.split("[–]",expand=True)
partidos.drop("Marcador",axis=1,inplace=True)
```

```python
partidos=partidos.astype({"Goles Local":int,"Goles Visitante":int})
```

```python
partidos.dtypes
```

```python
partidos["Goles Totales"]=partidos["Goles Local"]+partidos["Goles Visitante"]
```

```python
partidos.to_csv("Mundiales limpio.csv",index=False)
fixtures.to_csv("fixtures.csv",index=False)
```
